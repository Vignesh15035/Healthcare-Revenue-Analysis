```python
import pandas as pd

df = pd.read_csv("healthcare_dataset.csv")

df.head()
df.info()
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df["Date_of_Admission"] = pd.to_datetime(df["Date_of_Admission"], dayfirst=True, errors="coerce")
df["Discharge_Date"] = pd.to_datetime(df["Discharge_Date"], dayfirst=True, errors="coerce")
```

```python
df["Length_of_Stay"] = (df["Discharge_Date"] - df["Date_of_Admission"]).dt.days
```

```python
text_cols = ["Name","Doctor","Hospital","Gender","Admission_Type",
             "Medical_Condition","Insurance_Provider","Medication"]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
df["Gender"] = df["Gender"].replace({"M":"Male","F":"Female","male":"Male","female":"Female"})
```

```python
df["Billing_Amount"] = pd.to_numeric(df["Billing_Amount"], errors="coerce")
df["Billing_Amount"] = df["Billing_Amount"].fillna(df["Billing_Amount"].median())
```

```python
df["Date_of_Admission"] = pd.to_datetime(
    df["Date_of_Admission"], format="%Y-%m-%d", errors="coerce"
)
df["Discharge_Date"] = pd.to_datetime(
    df["Discharge_Date"], format="%Y-%m-%d", errors="coerce"
)
```

```python
df["Billing_Amount"] = df["Billing_Amount"].round(2)
```

```python
df.to_csv("healthcare_dataset_cleaned_file.csv", index=False)
```
